```python

Rock, Paper & Scissors

Let's play the famous game against our computer. You can check the rules here.

Task

Create a program that imitates the playability of the well known game of rock, paper, scissors. Follow the guidelines provided.

Tools

Loop: for/while
Functions: input(), print()...
Conditional statements: if, elif, else
Definition of functions. Modular programming
Import modules
To solve this challenge, the use of functions is recommended.

1. Import the choice function of the random module.
In [1]:
import random as rand
2. Create a list that includes the 3 possible gesture options of the game: 'rock', 'paper' or 'scissors'. Store the list in a variable called gestures.
In [ ]:
 
3. Create a variable called n_rounds to store the maximum number of rounds to play in a game.

Remember that the number of rounds must be odd: 1, 3, 5, ...

In [2]:
## NOTE: assumes user entered a non-negative integer -- TODO: add more input checking

def set_total_rounds():    
    n_rounds = input("Enter an odd number of rounds: ")
    n_rounds = int(n_rounds)

    ## make sure we've got an odd number
    if n_rounds % 2 == 0:
        n_rounds += 1
        print("Warning: You entered an even number. Setting the total number of rounds to %d" % n_rounds)

    return n_rounds
4. Create a variable called rounds_to_win to store the number of rounds that a player must win to win the game.

Hint: the value stored in rounds_to_win depends on the value of n_rounds.

In [3]:
def set_num_rounds_to_win(total_rounds):
    ## NOTE: assumes user entered a non-negative integer -- TODO: add more input checking

    rounds_to_win = input("Enter the number of rounds a player must win to win the game: ")
    rounds_to_win = int(rounds_to_win)

    ## make sure this number is not greater than the total number
    if rounds_to_win > total_rounds:
        print("Warning: You entered a value greater than the total number of rounds.")
        print("The number of rounds to win has been set to %d (i.e. *all* rounds)." % total_rounds)
        rounds_to_win = total_rounds

    return rounds_to_win
5. Create two variables to store the number of rounds that the computer and the player have won. Call these variables cpu_score and player_score.
In [4]:
# GLOBAL static list of possible gestures
# NOTE: ORDER OF ELEMENTS IN THIS LIST IS CRITICAL!!
gestures = ["rock", "paper", "scissors"]

# Global variables for designing winner status and for indices into scores list
TIE = 0
CPU = 1
PLAYER = 2

# for printing round / game outcome strings -- MUST agree with indices in the global vars above
player_names = ["tie", "cpu", "human"]
#scores = []

def print_scores(s):
    for i in range(len(s)):
        print("{}: {}".format(player_names[i], s[i]))
        
def initialize_player_scores(val=0):
    result = [val] * len(player_names)
    return result
6. Define a function that randomly returns one of the 3 gesture options.

You will use this function to simulate the gesture choice of the computer.

In [5]:
def cpu_throw(gesture_options):
    '''
    Select an item at random from the (global) gestures list
    Input: list of strings (gesture names)
    returns a string (the name of the gesture)
    '''
    result = rand.choice(gesture_options)
    return result
7. Define a function that asks the player which is the gesture he or she wants to show: 'rock', 'paper' or 'scissors'.

The player should only be allowed to choose one of the 3 gesture options. If the player's choice is not rock, paper or scissors, keep asking until it is.

In [6]:
def player_throw(gesture_options):
    '''
    Prompt user to enter their gesture for the next round
    NOTE: we're using input starting from 1 so the input options are close together on the user keyboard
    
    Input: list of strings (gesture names)
    returns a string (the name of the gesture)
    '''

    # construct prompt based on gestures list and store the list of valid user inputs
    user_prompt = ""
    valid_input = []
    
    for i in range(len(gesture_options)):
        my_num = i+1
        user_prompt += "Press %d for %s. " % (my_num, gesture_options[i])
        valid_input.append(str(my_num))
    user_prompt += "Go!"
    
    # continue prompting until they've entered a valid choice
    while True:
        player_throw_choice = input(user_prompt)
        if player_throw_choice not in valid_input:
            print("%s is not a valid option, try again" % player_throw_choice)
        else:
            break
            
    # now convert the user input number to their selected gesture string
    result = gesture_options[int(player_throw_choice) - 1]
    return result
8. Define a function that checks who won a round.

The function should return 0 if there is a tie, 1 if the computer wins and 2 if the player wins.

In [12]:
def get_winner(cpu_gesture, player_gesture, gesture_options):
    '''
    Determine the winner for this round. Uses numeric codes to describe the winner 
    (0 if tie, 1 if the cpu wins, 2 if the player wins)
    Input1: string (the name of the computer's gesture for this round)
    Input2: string (the name of the players's gesture for this round)
    Returns: int 
    
    '''
    
    player_gesture_i = gesture_options.index(player_gesture)
    cpu_gesture_i = gesture_options.index(cpu_gesture)
    
    if player_gesture_i == cpu_gesture_i:
        result = TIE
    elif player_gesture_i == (cpu_gesture_i +1) % len(gesture_options):
        result = PLAYER
    elif len(gesture_options) == 5 and player_gesture_i == (cpu_gesture_i +3) % 5:
        result = PLAYER
    else:
        result = CPU
    
    return result
    
9. Define a function that prints the choice of the computer, the choice of the player and a message that announces who won the current round.

You should also use this function to update the variables that count the number of rounds that the computer and the player have won. The score of the winner increases by one point. If there is a tie, the score does not increase.

In [8]:
def report_current_round(cpu_gesture, player_gesture, scores, gesture_options):
    '''
    Report which gestures where thrown by whom, and report (and track) who won this round
    Input1: string (the name of the computer's gesture for this round)
    Input2: string (the name of the players's gesture for this round)    
    Returns: int (numeric code describing winner of this round)
    '''
    
    winner_code = get_winner(cpu_gesture, player_gesture, gesture_options)
    scores[winner_code] += 1
    
    print("{} throws: {}".format(player_names[CPU], cpu_gesture))
    print("{} throws: {}".format(player_names[PLAYER], player_gesture))
    print("\t Round winner: {}\n".format(player_names[winner_code]))

    return winner_code, scores
    
10. Now it's time to code the execution of the game using the functions and variables you defined above.

First, create a loop structure that repeats while no player reaches the minimum score necessary to win and the number of rounds is less than the maximum number of rounds to play in a game.

Inside the loop, use the functions and variables above to create the execution of a round: ask for the player's choice, generate the random choice of the computer, show the round results, update the scores, etc.

In [9]:
def doitall(gesture_options):
    # initialize number of ties and wins to 0
    scores = initialize_player_scores()

    n_rounds = set_total_rounds()
    rounds_to_win = set_num_rounds_to_win(n_rounds)

    print("\nOK, the winner must win {}/{} rounds to win the game... Begin!".format(rounds_to_win, n_rounds))

    current_round = 1

    while current_round <= n_rounds:
        print("Round {}... fight!".format(current_round))

        p = player_throw(gesture_options)
        c = cpu_throw(gesture_options)
        winner_code, scores = report_current_round(c, p, scores, gesture_options)

        if winner_code > 0 and scores[winner_code] == rounds_to_win:
            print("{} has won {} rounds in total, and has WON THE GAME!".format(player_names[winner_code], scores[winner_code]))
            break
#         else: #uncomment else block for debugging
#             print("{} has won {} rounds in total, which is not enough to win the game".format(player_names[winner_code], scores[winner_code]))

        # get ready for the next loop
        current_round += 1

    # handle the case where the final winning condition was not met
    if current_round > n_rounds:
        print("The game ended without an overall winner")
        if scores[CPU] > scores[PLAYER]:
            print("However, {} won more rounds.".format(player_names[CPU]))
        elif scores[CPU] < scores[PLAYER]:
            print("However, {} won more rounds.".format(player_names[PLAYER]))

    print("\nFinal score:")
    print_scores(scores)
11. Print the winner of the game based on who won more rounds.

Remember that the game might be tied.

In [10]:
doitall(gestures)
Enter an odd number of rounds: 5
Enter the number of rounds a player must win to win the game: 2

OK, the winner must win 2/5 rounds to win the game... Begin!
Round 1... fight!
Press 1 for rock. Press 2 for paper. Press 3 for scissors. Go!3
cpu throws: scissors
human throws: scissors
	 Round winner: tie

Round 2... fight!
Press 1 for rock. Press 2 for paper. Press 3 for scissors. Go!1
cpu throws: scissors
human throws: rock
	 Round winner: human

Round 3... fight!
Press 1 for rock. Press 2 for paper. Press 3 for scissors. Go!5
5 is not a valid option, try again
Press 1 for rock. Press 2 for paper. Press 3 for scissors. Go!2
cpu throws: scissors
human throws: paper
	 Round winner: cpu

Round 4... fight!
Press 1 for rock. Press 2 for paper. Press 3 for scissors. Go!2
cpu throws: paper
human throws: paper
	 Round winner: tie

Round 5... fight!
Press 1 for rock. Press 2 for paper. Press 3 for scissors. Go!1
cpu throws: paper
human throws: rock
	 Round winner: cpu

cpu has won 2 rounds in total, and has WON THE GAME!

Final score:
tie: 2
cpu: 2
human: 1
Bonus: Rock, Paper, Scissors, Lizard & Spock



In this challenge, you need to improve the previous game by adding two new options. To know more about the rules of the improved version of rock, paper, scissors, check this link.

In addition, you will also need to improve how the game interacts with the player: the number of rounds to play, which must be an odd number, will be requested to the user until a valid number is entered. Define a new function to make that request.

Hint: Try to reuse the code that you already coded in the previous challenge. If your code is efficient, this bonus will only consist of simple modifications to the original game.

In [14]:
five_gestures = ["rock", "paper", "scissors", "spock", "lizard"]

doitall(five_gestures)
Enter an odd number of rounds: 7
Enter the number of rounds a player must win to win the game: 5

OK, the winner must win 5/7 rounds to win the game... Begin!
Round 1... fight!
Press 1 for rock. Press 2 for paper. Press 3 for scissors. Press 4 for spock. Press 5 for lizard. Go!5
cpu throws: paper
human throws: lizard
	 Round winner: human

Round 2... fight!
Press 1 for rock. Press 2 for paper. Press 3 for scissors. Press 4 for spock. Press 5 for lizard. Go!5
cpu throws: rock
human throws: lizard
	 Round winner: cpu

Round 3... fight!
Press 1 for rock. Press 2 for paper. Press 3 for scissors. Press 4 for spock. Press 5 for lizard. Go!3
cpu throws: rock
human throws: scissors
	 Round winner: cpu

Round 4... fight!
Press 1 for rock. Press 2 for paper. Press 3 for scissors. Press 4 for spock. Press 5 for lizard. Go!4
cpu throws: scissors
human throws: spock
	 Round winner: human

Round 5... fight!
Press 1 for rock. Press 2 for paper. Press 3 for scissors. Press 4 for spock. Press 5 for lizard. Go!4
cpu throws: lizard
human throws: spock
	 Round winner: cpu

Round 6... fight!
Press 1 for rock. Press 2 for paper. Press 3 for scissors. Press 4 for spock. Press 5 for lizard. Go!3
cpu throws: rock
human throws: scissors
	 Round winner: cpu

Round 7... fight!
Press 1 for rock. Press 2 for paper. Press 3 for scissors. Press 4 for spock. Press 5 for lizard. Go!2
cpu throws: scissors
human throws: paper
	 Round winner: cpu

cpu has won 5 rounds in total, and has WON THE GAME!

Final score:
tie: 0
cpu: 5
human: 2
```
